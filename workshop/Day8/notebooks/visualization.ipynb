```python
# Generating data
x = list(range(10))
y = list(range(20,30))
```

```python
image = np.random.randint(1, size=(20,20)) # generating random image
image[5:10,5:10] = 1
image = plt.imread('sulatn.jpg')
```

```python
plt.clf()
figure,locations = subplots(2,2)
locations[0,1].imshow(image)
locations[0,1].set_title("sultan")
locations[0,0].scatter(x,y)
locations[0,0].set_title("Scatter")
locations[1,0].bar(x,x)
locations[1,1].plot(y)
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_excel('BreastTissue.xls','Data')
df['Class'] = pd.Categorical(df.Class)
df.describe()
```

```python
plt.figure()
count = df.groupby(by="Class").count()
count.plot(kind='bar',stacked=True)
plt.show()
```

```python
df['Class'].value_counts().plot(kind='bar')
plt.show()
```

```python
df['Class'].value_counts().plot(kind='pie')
plt.show()
```

```python
from pandas.plotting import scatter_matrix
plt.clf()
scatter_matrix(df.iloc[:,1:], alpha=1, figsize=(10, 10), diagonal='kde')
plt.show()
```

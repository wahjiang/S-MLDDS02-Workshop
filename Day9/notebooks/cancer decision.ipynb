```python
from sklearn.utils import shuffle
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_excel("BreastTissue.xls","Data")
dataset = dataset.drop("Case #",axis=1)
dataset = shuffle(dataset)
```

```python
dataset.describe()
```

```python
dataset.head()
```

```python
labels = dataset.iloc[:,0]
attr = dataset.iloc[:,1:]
attr=(attr-attr.min())/(attr.max()-attr.min())
n = 75
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0, splitter="best")

clf.fit(attr.iloc[:n,:],labels.iloc[:n])
y_predicted = clf.predict(attr.iloc[n+1:,:])

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

print "Accuracy = {} %".format(accuracy_score(labels.iloc[n+1:], y_predicted)*100)
print "Classification Report \n {}".format(classification_report(labels.iloc[n+1:], y_predicted, labels=dataset.Class.unique()))
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()

clf.fit(attr.iloc[:n,:],labels.iloc[:n])
y_predicted = clf.predict(attr.iloc[n+1:,:])

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

print "Accuracy = {} %".format(accuracy_score(labels.iloc[n+1:], y_predicted)*100)
print "Classification Report \n {}".format(classification_report(labels.iloc[n+1:], y_predicted, labels=dataset.Class.unique()))
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()

clf.fit(attr.iloc[:n,:],labels.iloc[:n])
y_predicted = clf.predict(attr.iloc[n+1:,:])

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

print "Accuracy = {} %".format(accuracy_score(labels.iloc[n+1:], y_predicted)*100)
print "Classification Report \n {}".format(classification_report(labels.iloc[n+1:], y_predicted, labels=dataset.Class.unique()))
```

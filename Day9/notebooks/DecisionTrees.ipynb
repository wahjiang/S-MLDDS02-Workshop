```python
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(splitter="random")
iris = load_iris()
```

```python
import numpy as np
np.mean(iris.data[:,3])
```

```python
clf = clf.fit(iris.data, iris.target)
tree.export_graphviz(clf,out_file='tree_iris.dot')
```

```python
from graphviz import Source
from sklearn import tree
Source( tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names))
```
